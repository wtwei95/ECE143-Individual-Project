# Heat Map

```python
def heatMap(dataFrame):
    figList = []
    for i in range(24):
        if i < 10:
            df = dataFrame['2016-01-01 0' + str(i)]
        else:
            df = dataFrame['2016-01-01 ' + str(i)]
        geo = geoInfo(df.Pickup_latitude,df.Pickup_longitude)[:500]
        fig = gmaps.figure()
        fig.add_layer(gmaps.heatmap_layer(geo))
        figList.append(fig)
    return figList
```

```python
def geoInfo(latitude,longitude):
    p_lati = []
    p_long = []
    for pl in latitude:
        p_lati.append(pl)
    for pl in longitude:
        p_long.append(pl)
    return zip(p_lati,p_long)
```
